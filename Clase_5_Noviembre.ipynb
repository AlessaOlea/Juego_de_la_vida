Mathplotlib y seaborn sirven para la represntacion visual por medio de graficas de los datos o informacion obtenida en el codiigo.
Seaborn es un poco mas elegante que Mathplotlib

```python
import matplotlib.pyplot as plt #Importamos las bibliotecas
import numpy as np
```

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints) #Cosas a imprimir
plt.show() #Imprimidor
```

```python
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

#plt.plot(x,y1,y2,"o")
plt.plot(x,y1,"o") #Lo que est√° en comillas camnbia la forma o puntos de la grafica
plt.plot(x,y2,"*")
plt.show()
```

```python
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

#plt.plot(x,y1,y2,"o")

plt.plot(x,y1,"r") #la letra entre comillas cambia el color de la grafica
plt.plot(x,y2,"g")
plt.show()
```

```python
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

#plt.plot(x,y1,y2,"o")
plt.plot(x,y1,marker = "o") #"Marker" une los puntos
plt.plot(x,y2,marker = "*")
plt.show()
```

```python
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

#plt.plot(x,y1,y2,"o")

plt.grid(color = 'gray', linestyle = '--', linewidth = 2)

plt.plot(x,y1,"o") #Lo que est√° en comillas camnbia la forma o puntos de la grafica
plt.plot(x,y2,"*")
plt.show()
```

```python
x = np.linspace(-20,20,100)
y1 = x**3

plt.grid()
plt.plot(y1,'o-k' , ms = 20, mec = 'r')
plt.show()
```

```python
x = np.linspace(-20,20,100)
y1 = x**3

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
font3 = {'family':'serif','color':'magenta','size':15}

plt.xlabel("x", fontdict = font1)
plt.ylabel("y", fontdict = font2)
plt.title("La gr√°fica de x^2", fontdict = font3)

plt.grid(axis = 'x')
plt.plot(y1, color = 'm', ls = "-", linewidth = "5.5")
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, color = 'm', ls = "-", linewidth = "5.5")
plt.plot(y2, color = 'r', ls = "-", linewidth = "5.5")

plt.xlabel("Lo que me debe Paco")
plt.ylabel("Lo que me debe Andres")
plt.title("Los que me deben")

plt.grid(axis = 'y')
plt.show()
```
